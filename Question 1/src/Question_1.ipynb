# Question 1 : Frailty Analysis

```python
import pandas as pd
```

```python
df=pd.read_csv("/content/Project/data_raw/frality_raw_data.csv")
```

```python
df.head()
```

```python
df["height_m"] = df["height"] * 0.0254
df["weight_kg"] = df["weight"] * 0.45359237
df.head()
```

```python
df["BMI"] = (df["weight"] / (df["height"]**2)).round(2)
df
```

```python
df["age_group"] = ""


df.loc[df["age"] < 30, "age_group"] = "<30"
df.loc[(df["age"] >= 30) & (df["age"] <=45), "age_group"] = "30-45"
df.loc[(df["age"]>=46) & (df["age"] <=60), "age_group"] = "40-60"
df.loc[df["age"] >60, "age_group"] = ">60"
df
```

```python
df["frailty_binary"] = df["frailty"].map({"Y": 1, "N": 0}).astype("int8")
df
```

```python
df = pd.get_dummies(df, columns=['age_group'])
df
```

```python
summary = df.describe().T[["mean", "std"]]
summary["median"] = df.median(numeric_only=True)

summary
```

```python
import os
os.makedirs("/content/Project/reports", exist_ok=True)
summary.to_markdown("/content/Project/reports/findings.md")
```

```python
correlation = df["grip_strength"].corr(df["frailty_binary"])
correlation
```

```python
with open("/content/Project/reports/findings.md", "a") as f:
    f.write("\n\n## Correlation Analysis\n")
    f.write(f"\nCorrelation between grip_strength and frailty_binary: {correlation:.4f}\n")
```
